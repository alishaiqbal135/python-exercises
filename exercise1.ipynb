```python
#Q1: String operations
def string_operations(s):
    reversed_str = s[::-1]
    every_second = s[::2]
    return reversed_str, every_second

# Example
print(string_operations("hello"))
```

```python
#Q2: Squares of even numbers
def square_even(nums):
    return [x**2 for x in nums if x % 2 == 0]

print(square_even([1,2,3,4,5,6]))
```

```python
#Q3: Character frequency
def char_frequency(s):
    freq = {}
    for ch in s:
        freq[ch] = freq.get(ch, 0) + 1
    return freq

print(char_frequency("hello"))
```

```python
#Q4: Indices of odd numbers
def odd_indices(lst):
    return [i for i, x in enumerate(lst) if x % 2 != 0]

print(odd_indices([1,2,3,4,5]))
```

```python
#Q5:Rotate list right by k
def rotate_list(lst, k):
    k = k % len(lst)
    return lst[-k:] + lst[:-k]

print(rotate_list([1,2,3,4,5], 2))
```

```python
#Q6: Two lists to dictionary
def list_to_dict(keys, values):
    return dict(zip(keys, values))
print(list_to_dict(['a','b','c'], [1,2,3]))
```

```python
#Q7: iter() and next()
def manual_iteration(lst):
    it = iter(lst)
    while True:
        try:
            print(next(it))
        except StopIteration:
            print("Iteration finished")
            break

manual_iteration([10,20,30])
```

```python
#Q8: Flatten nested list
def flatten_list(nested):
    return [item for sublist in nested for item in sublist]

print(flatten_list([[1,2],[3,4],[5]]))
```

```python
#Q9: Palindrome check
def is_palindrome(s):
    return s == s[::-1]

print(is_palindrome("madam"))
```

```python
#Q10: Second largest (without sorting)
def second_largest(lst):
    first = second = float('-inf')
    
    for num in lst:
        if num > first:
            second = first
            first = num
        elif num > second and num != first:
            second = num
            
    return second

print(second_largest([10,20,4,45,99]))
```
